Import the necessary libraries and create your dataframe(s).

```python
# Import the appropriate libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

# Create dataframe from csv
life_df = pd.read_csv("Human_life_Expectancy.csv")
```

```python
# replace all 'Not Available' values with NaN, which is float
life_df = life_df.replace('Not Available', np.nan)

# change dtype of all year columns to float
life_df = life_df.astype({"1990": float})
life_df = life_df.astype({"1991": float})
life_df = life_df.astype({"1992": float})
life_df = life_df.astype({"1993": float})
life_df = life_df.astype({"1994": float})
life_df = life_df.astype({"1995": float})
life_df = life_df.astype({"1996": float})
life_df = life_df.astype({"1997": float})
life_df = life_df.astype({"1998": float})
life_df = life_df.astype({"1999": float})
life_df = life_df.astype({"2000": float})
life_df = life_df.astype({"2001": float})
life_df = life_df.astype({"2002": float})
life_df = life_df.astype({"2003": float})
life_df = life_df.astype({"2004": float})
life_df = life_df.astype({"2005": float})
life_df = life_df.astype({"2006": float})
life_df = life_df.astype({"2007": float})
life_df = life_df.astype({"2008": float})
life_df = life_df.astype({"2009": float})
life_df = life_df.astype({"2010": float})
life_df = life_df.astype({"2011": float})
life_df = life_df.astype({"2012": float})
life_df = life_df.astype({"2013": float})
life_df = life_df.astype({"2014": float})
life_df = life_df.astype({"2015": float})
life_df = life_df.astype({"2016": float})
life_df = life_df.astype({"2017": float})
life_df = life_df.astype({"2018": float})
life_df = life_df.astype({"2019": float})

#life_df.dtypes
```

## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

### Notes
- too many missing values to search/replace actual info UNLESS you limit the missing values to search/replace
- for empty national values, cannot replace with different year values or avg existing subnational values
- for empty subnational values, cannot replace with different year values or national value
- need to pick if dropping certain countries/regions for all years OR dropping certain years for all countries for clean dataset
- Replacement values from https://data.worldbank.org/indicator/SP.DYN.LE00.IN

```python
# Are the same countries missing values for most/all years?

# VISUALIZATION 2 - cont. from EDA
# histogram showing the number of subnations for each country
# commented out code for all countries graph
# showing code for ONLY countries w/ null values graph

figure(figsize=(40, 8), dpi=80)

countries = pd.unique(life_df['Country'])
country_counts = []
null_countries = []
total_nulls = [] # used for VIS.3 - total null value counts per country over all years
# nulls_peryear = [] # used for VIS.4

for x in countries:
    full_rows = life_df.loc[life_df['Country'] == x]
    # country_counts.append(len(full_rows)-1)
    
    sum_nulls = sum(list(full_rows.isna().sum())) # used for VIS.3
    
    if sum_nulls > 0: # filtered for countries that have null values
        null_countries.append(x)
        country_counts.append(len(full_rows)-1) # remove nation from subnations count
        total_nulls.append(sum_nulls) # used for VIS.3
        # nulls_peryear.append(list(full_rows.isna().sum())) # used for VIS.4

# creating dataframe with values from for-loop
nullcountry_df = pd.DataFrame(
    {'Name': null_countries,
     # 'Name': countries,
     'Subnation_Count': country_counts,
     'Null_Count': total_nulls
    })

nullcountry_df.sort_values(by=['Subnation_Count'], ascending=False, inplace=True)

plt.rcParams['font.size'] = '30'
fig2 = sns.barplot(x=list(nullcountry_df['Name']),y=list(nullcountry_df['Subnation_Count']))
fig2.set(xlabel='Countries w/ Null Values', ylabel='# Subnations', title = "# Subnations per Country")
plt.xticks(rotation=90)

plt.show()
print(str(len(null_countries)) + " Countries have NaN values")
#print(null_countries)
#print(total_nulls)
```

```python
# VISUALIZATION 3: bar graph comparing total null value counts per country over all years
# using data from VIS.2 code cell

figure(figsize=(40, 8), dpi=80)

nullcountry_df.sort_values(by=['Null_Count'], ascending=False, inplace=True)

plt.rcParams['font.size'] = '30'
fig3 = sns.barplot(x=list(nullcountry_df['Name']),y=list(nullcountry_df['Null_Count']))
fig3.set(xlabel='Countries w/ Null Values', ylabel='# Nulls', title = "# Nulls per Country")
plt.xticks(rotation=90)

plt.show()
# print(nullcountry_df.loc[nullcountry_df['Null_Count'] < 50])
```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
# checking that the Country and Country_Code columns are 1 to 1
# the counts of country name and country code should be the same

country_series = clean_df['Country'].value_counts()
code_series = clean_df['Country_Code'].value_counts()
print(country_series.isin(code_series).describe())
```

```python
# checking that 'Total' values under Region column correspond to
#  'National' values under Level column
one_to_one = True

for x in range(len(clean_df)):
    row = clean_df.iloc[x,:]
    if ((row['Level']== 'National' and row['Region'] != 'Total')
        or (row['Level']!= 'National' and row['Region'] == 'Total')):
        one_to_one = False

print(one_to_one)
```

```python
# checking that national value is approx avg of subnational values

#Import math Library
import math 
import statistics

countries = pd.unique(clean_df['Country'])
years = clean_df.iloc[:,4:33]
approx = True # whether expectation is true

for col in years.columns: # for each year column
    for x in countries: # for each country name value
        country_group = list(clean_df.loc[(clean_df['Country'] == x)
                                         & (clean_df['Level'] == 'subnational')][col])
        
        total = clean_df.loc[(clean_df['Country'] == x)
                                         & (clean_df['Level'] == 'National')][col]
        if len(country_group) > 0:
            avg = statistics.mean(country_group)
            if (not(math.isclose(total, avg, rel_tol = 0.5))):
                approx = False

print(approx)
```
